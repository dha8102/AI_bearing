# IRIS 데이터를 이용한 회귀 분석

[IRIS 데이터셋] \
통계학과 머신러닝에서 자주 사용되는 데이터셋. \
붓꽃(iris)의 품종을 분류하기 위한 데이터셋. \
총 세 가지 품종 (Setosa, Versicolor, Virginica)의 정보가 각각 50개씩 총 150개의 샘플로 구성됨

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Iris.csv')
df
```

[데이터셋 설명]\
Id : 인덱스\
SepalLengthCm : 꽃받침의 길이(cm)\
SepalWidthCm : 꽃받침의 너비(cm)\
PetalLengthCm : 꽃잎의 길이(cm)\
PetalWidthCm : 꽃잎의 너비(cm)\
Species : 종

```python
df.info()
```

```python
df.describe()
```

```python
# matplotlib
for i in df.columns:
    plt.hist(df[i])
    plt.xticks()
    plt.xlabel(i)
    plt.ylabel('counts')
    plt.show()
```

```python
# matplotlib
fig, axes = plt.subplots(2, 2, figsize = (10,10))
col_cnt = 1
col_names = df.columns.tolist()
for name in col_names:
  if name != 'Id' and name != 'Species':
    plt.subplot(2, 2, col_cnt)
    plt.boxplot(df[name])
    plt.title(name)
    col_cnt += 1
```

```python
sns.pairplot(df, diag_kind="kde");
```
